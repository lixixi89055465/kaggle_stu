```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
sns.set_style('darkgrid') 
sns.set_context('paper') 
# 设置风格、尺度 
import warnings
warnings.filterwarnings('ignore') 
# 不发出警告
```

```python
# 基本用法 
tips= pd.read_csv('tips.csv') 
print(tips.head()) 
```

```python
sns.lmplot(
    x='total_bill',y='tip',
#           hue='smoker',
           data=tips,
            palette='Set1',
           ci=70,#误差值 
#            size=5,#图表大小 
#            markers=['+','o'],# 点样式            
          )
```

```python
# 拆分多个表格 
sns.lmplot(x='total_bill',y='tip',col='smoker',data=tips)
```

```python
# 多图标1
sns.lmplot(x='size',y='total_bill',hue='day',col='day',data=tips, 
          aspect=0.6,# 长宽比 
           x_jitter=.6, # 给X或者Y 轴随机增加噪声点 
           col_wrap=2,#每行的列数 
          
          )
```

```python
'''
多图表 2 
'''
sns.lmplot(x='total_bill',y='tip',row='sex',col='time',data=tips
          ) 
# 行为sex字段，列为time字段
# x轴total_bill，y轴tip
```

```python
'''
非线形回归 
'''

sns.lmplot(x='total_bill',y='tip',data=tips, order=2 ) 
```

```python
'''
非线形回归 
'''

sns.lmplot(x='total_bill',y='tip',data=tips, order=3) 
```
