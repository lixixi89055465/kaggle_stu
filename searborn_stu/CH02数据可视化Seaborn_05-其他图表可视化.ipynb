```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
sns.set_style('darkgrid') 
sns.set_context('paper') 
# 设置风格、尺度 
import warnings
warnings.filterwarnings('ignore') 
# 不发出警告
```

```python
'''
1.时间线图表-tsplot() 
简单示例 
'''
x=np.linspace(0,15,31) 
data=np.sin(x) + np.random.rand(10,31) + np.random.randn(10,1) 
print(data.shape) 
print(pd.DataFrame(data).head())
```

```python

sns.tsplot(data=data,
           err_style='ci_band',
           interpolate=True,
#            ci=[40,70,90],#设置误差区间 
           color='g' #设置颜色 
          ) 
```

```python
sns.tsplot(data=data,
            err_style='boot_traces',
           interpolate=False,
#            ci=[30,40,70,90],#设置误差区间 
           color='g' #设置颜色 
          ) 
pd.DataFrame(data).head()
```

```python
sns.tsplot(data=data,err_style='boot_traces', 
          n_boot=300# 迭代次数 
          ) 
```

```python
'''
1.时间线图表  - tsplot() 
参数设置  
'''
gamma= pd.read_csv('gammas.csv')
print(gamma.head()) 
```

```python

print("数据量为:%i条 "%len(gamma)) 
print("timepoint w为 0.0时的数据量为:%i条 "%len(gamma[gamma['timepoint']==0]))
print('timepoint 共有%i个唯一值 '%len(gamma['timepoint'].value_counts()))
print(gamma['timepoint'].value_counts()) # 查看唯一具体信息 
# 导入数据 
```

```python
'''
2.热图 -heatmap() 
简单示例 
'''
df = pd.DataFrame(np.random.rand(10,12)) 
# 创建数据  - 10*12 图表 
df 
```

```python
'''
2.热图  - heatmap ()  
参数设置 


'''
flights= pd.read_csv('flights.csv')
flights= flights.pivot('month','year','passengers') 
sns.heatmap(flights,
           annot=True, #是否显示数值 
            fmt='d',# 格式化字符串 
            linewidths=0, # 格子边线宽度 
            # center=100,# 调色盘的色彩中心值，若没有指定，则以cmap 为主 
            # cmap='Redis',# 设置调色盘 
            cbar=True,# 是否显示图例色带 
#             cbar_kws={'orientation':'horizontal'},#是否横向显示图例色带 
            # square=True,#是否正方形显示图表 
           )
flights.head() 
```

```python
'''
2.热图 - heatmap() 
绘制半边热图 

'''
sns.set(style='white') 
# 设置风格 
rs= np.random.RandomState(33) 
d= pd.DataFrame(rs.normal(size=(100,26)))
corr= d.corr() # 求解相关性矩阵表格 
# 创建数据 
d.head() 
```

```python
corr.head()
```

```python
mask= np.zeros_like(corr,dtype=np.bool) 
mask[np.triu_indices_from(mask)]=True 
# 设置一个’上三角形“蒙版 

sns.diverging_palette
```

```python
# cmap= sns.diverging_palette(220,10,as_cmap=True) 
# 设置调色盘 
sns.heatmap(corr,mask=mask,
#             cmap=cmap,
            vmax=.9,center=1, 
           square=True,linewidths=0.2 )

# 生成半边热图 
```
