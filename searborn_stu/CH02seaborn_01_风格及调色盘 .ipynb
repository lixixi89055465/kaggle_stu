```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*(7-i)*flip))
sinplot()
#  
```

```python
sns.set()
sinplot()
plt.grid(linestyle='--')
```
