```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 创建正弦函数及图标
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*(7-i)*flip))
sinplot()
# 创建正弦函数，及出图 
```

```python
sns.set()
sinplot()
plt.grid(linestyle='--')
```

```python
'''
2.set_style()
切换seaborn 图表分格 
风格选择包括：'white','dark','whitegrid','darkgrid','ticks'
'''
```

```python
fig=plt.figure(figsize=(6,6))
ax1=fig.add_subplot(2,1,1)
# sns.set_style('whitegrid')
sns.set_style('ticks')
data= np.random.normal(size=(20,6))+np.arange(6)/2 
print(data.shape)
sns.boxplot(data=data) 
plt.title('style- whitegrid') 
# 仍然可以使用matplotlib 的参数 
ax2=fig.add_subplot(2,1,2)
sinplot()
```

```python
'''
3. despine()
设置图标坐标轴 
seaborn.despine()
'''
sns.set_style('ticks')
# 设置风格 
fig=plt.figure(figsize=(6,9))
plt.subplots_adjust(hspace=0.3 )
# 创建图表 
ax1=fig.add_subplot(3,1,1)
sinplot()
sns.despine()

# 删除了上、右坐标轴 
```

```python

ax2=fig.add_subplot(3,1,2) 
sns.violinplot(data=data) 
sns.despine(offset=10,trim=False)
# offset: 与坐标轴之间的偏移 
# trim: 为true 时，将坐标轴限制在数据最大最小值 
```

```python
ax3=fig.add_subplot(3,1,3)
sns.boxplot(data=data,palette='deep') 
sns.despine(left=True,right=False)
# top , right ,left, bottom:布尔型，为true 时不显示 
```

```python
'''
4.axes_style()
设置局部图标风格，可学习和 with 配合的用法 

'''
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
# 设置局部图标风格，用with 做代码块区分 
# sns.set_style('whitegrid')
plt.subplot(212)
sinplot()
# 外部表格风格 
```

```python
'''
5. set_context() 
设置显示比例尺度 
选择包括:'paper','notebook','talk','poster' 

'''
# sns.set_context('paper') 
# sns.set_context('notebook') 
sns.set_context('talk') 
sinplot()
# 默认为notebook 
```

```python
# 分组颜色设置 - "Paired"
sns.palplot(sns.color_palette('Paired',20))
```

```python
'''
2.设置亮度，饱和度 
可用方法:
2.1 husl_palette([n_colors,h,s,l])
2.2 hls_palette([n_colors,h,l,s])
l -> 亮度 
2 -> 饱和度 

'''
# sns.palplot(sns.hls_palette(8,l=.9,s=.8 ))
sns.palplot(sns.husl_palette(8,l=.5,s=1))
```

```python
'''
3.cubehelix_palette() 
 按照线性增长计算，设置颜色 
 '''
sns.palplot(sns.cubehelix_palette(8,gamma=2))
sns.palplot(sns.cubehelix_palette(8,start=1.2,rot=-.1))
sns.palplot(sns.cubehelix_palette(8,start=1.2,rot=-.5))
sns.palplot(sns.cubehelix_palette(8,start=2,rot=0,dark=0,light=.95,reverse=True))
'''
n_colors -> 颜色个数 
start-> 值区间在 0-3 ,开始颜色 
rot-> 颜色旋转 角度 
gamma -> 颜色gamma值 ，越大颜色越暗 
dark,light -> 值区间0-1，颜色深浅 
reverse -> 布尔值，默认为 False，由浅到深 
'''
```

```python
'''
4. dark_palette(color,[,n_colors,reverse,...]) /light_palette(color[,n_colors,reverse,...])
颜色深度 

'''
sns.palplot(sns.light_palette('green')) # 按照green 做浅色调色盘 
```

```python
sns.palplot(sns.color_palette('Greens')) # cmap 为green 风格 
```

```python
sns.palplot(sns.dark_palette('red',reverse=True))# 按照blue 做深色 调色盘 
# reverse -> 转制颜色 
```

```python
sns.palplot(sns.diverging_palette(145,280,s=85,l=25,n=7,center='light'))
'''
h_neg,h_pos -> 起始/终止颜色值 
s -> 值区间0-100，饱和度 
l -> 值区间0-100, 亮度 
n -> 颜色个数 
center -> 中心颜色为浅色还是深色 ‘light','dark',默认为light 
'''
```
