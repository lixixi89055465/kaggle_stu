```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
sns.set_style('whitegrid')
sns.set_context('paper') 
# 设置风格、尺度 
import warnings
warnings.filterwarnings('ignore') 
#不发出警告 
```

```python
'''
1.stripplot() 
按照不同类别对样本数据进行分布散点图绘制 

'''
tips= pd.read_csv('tips.csv') 
#加载数据
tips.head()
```

```python
sns.stripplot(x='day', # x-> 设置分组统计字段 
             y='total_bill', # y-> 数据分布统计字段 
              #这里xy数据对调，将会使得散点图横向分布 
              data=tips, # data - > 对应数据 
              jitter=True , # jitter -> 当点数据重合较多时，用该参数做一些跳转，也可以设置间距如:jitter =0.1 
#               jitter=False,
              size=5,edgecolor='w',linewidth=1,marker='o' #设置点的大小 、描边颜色或宽度、点样式 

             )
```

```python
print(tips['day'].value_counts())
tips.head()
```

```python
'''
1.stripplot() 
通过hue 参数再次分类 
'''

sns.stripplot(x='sex',y='total_bill',hue='day', data=tips,jitter=True) 
```

```python
'''
1.stripplot() 
设置调色盘 
'''
sns.stripplot(x='sex',y='total_bill',hue='day',
             data=tips,jitter=True, palette='Set2',# 设置调色盘 
#               dodge=True,# 是否拆分 
              dodge=False,
             )
```

```python
#分簇散点图 
sns.swarmplot(x='total_bill',y='day',data=tips, 
             size=5,edgecolor='w',linewidth=1,marker='o', 
             palette='Reds')
```

```python
#分簇散点图 
sns.swarmplot(x='day',y='total_bill',data=tips, 
             size=5,edgecolor='w',linewidth=1,marker='o', 
             palette='Reds')
```

```python
'''
1.stripplot() 
#筛选分类类别 

'''
print(tips['day'].value_counts())
# 查看day 字段的唯一值 
sns.swarmplot(x='day',y='total_bill',data=tips, order=['Sat','Sun'])
```

```python
'''
1.boxplot() 
箱型图
'''
sns.boxplot(x='day',y='total_bill',data=tips, 
           linewidth=2,# 线宽 
            width=0.5,# 箱之间的间隔比例 
            fliersize=3,# 异常点的大小 
            palette='hls',
            whis=1.5,# 设置IQR 
            notch=True,#设置是否以中值做凹槽 
            order=['Thur','Fri','Sat','Sun'],# 筛选类别 
           )
#绘制箱型图 
sns.swarmplot(x='day',y='total_bill',data=tips,color='k',size=3,alpha=0.8 )
#可以添加散点图 
```

```python
# 1.boxplot() 
# 通过hue参数再分类 
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker',palette='Reds') 
# 绘制箱型图 

sns.swarmplot(x='day',y='total_bill',data=tips,color='k',size=3,alpha=0.8 )
# 可以添加散点图 
```

```python
'''
2.violinplot() 
小提琴图 

'''
sns.violinplot(x='day',y='total_bill',data=tips, 
              linewidth=2,#线宽 
               width=0.8,# 箱之间的间隔比例 
               palette='hls',# 设置调色板 
               order=['Thur','Fri','Sat','Sun'],# 筛选类别  
               scale='count',# 测试小提琴图的宽度:area-面积相同，count-按照样本数量决定宽度
               gridsize=100,# '测试xi小提琴边线的平滑度 ，越高约平滑 '
               inner='box',#设置内部显示类型，-> 'box,quartile , point,stick,None'
               bw=0.2 # 控制拟合程度，一般可以不设置 
              )
# 用法和boxplot类似 
```

```python
# 2. violinplot() 
# 通过hue参数再分类 
sns.violinplot(x='day',y='total_bill',data=tips, 
              hue='smoker',palette='muted',
               split=False,# 设置是否拆分小提琴图 
               inner='quartile'
              )
```

```python
'''
2.violinplot() 
结合散点图 

'''
sns.violinplot(x='day',y='total_bill',data=tips,palette='hls',inner=None)
sns.swarmplot(x='day',y='total_bill',data=tips,color='w',alpha=.5) 
# 插入散点图 
```

```python
'''
3.lvplot() 
LV 图表 
'''
sns.boxenplot(x='day',y='total_bill',data=tips,palette='mako',
          # hue ='smoker',
           width=0.8,# 箱之间间隔比例 
           linewidth=2, 
           scale='exponential',# 设置框的大小 -> 'linear','exponential','area'
           k_depth='trustworthy',# 设置框的数量-> 'proportion','tukey','trustworthy'
           )
# 绘制LV图表 

sns.swarmplot(x='day',y='total_bill',data=tips,color='k',size=3,alpha=0.8) 
# 可以添加散点图 
```

```python
'''
2.8 分类数据可视化 - 统计图 
 barplot() / countplot() / pointplot()  
 
 1.barplot()  
 柱状图 - 置信区间估计 
 置信区间  ：样本均值 + 抽样误差 
'''
titantic= pd.read_csv('titanic.csv')
print(titantic.head()) 
print("---------")
#加载数据 
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=titantic, 
           palette='hls',
            order=['male','female'],#筛选类别 
            capsize=0.05 ,# 误差线横向延申宽度 
            saturation=.8,# 颜色饱和度
            errcolor='k',
            errwidth=2,# 误差颜色，宽度 
            ci='sd'# 置信区间误差-》0-100内值，'sd',None
            )
```

```python
print(titantic.groupby(['sex','class']).mean()['survived']) 
print(titantic.groupby(['sex','class']).std()['survived'])
```

```python
'''
1、barplot() 
柱状图 - 置信区间估计 

'''
sns.barplot(x='day',y='total_bill',hue='sex',data=tips,
           palette='Blues',edgecolor='w')
tips.groupby(['day','sex']).mean() 
# 计算数据 
```

```python
'''
1.barplot()
柱状图 - 置信区间估计 

'''
crashes=pd.read_csv('car_crashes.csv')
crashes= crashes.sort_values('total',ascending=False) 
print(crashes.head())
```

```python
f,ax=plt.subplots(figsize=(6,15)) 
#创建图表 

sns.set_color_codes('pastel') 
sns.barplot(x='total',y='abbrev',data=crashes,label='Total',
           color='b',edgecolor='w')
# 设置一个柱状图 


sns.set_color_codes('muted') 
sns.barplot(x='alcohol',y='abbrev',data=crashes, 
           label='alcohol-involved',color='b',edgecolor='w') 
# 设置第二个柱状图 

ax.legend(ncol=2,loc='lower right') 
sns.despine(left=True,bottom=True) 
```

```python
'''
2.countplot() 
计数柱状图  

'''
sns.countplot(x='class',hue='who',data=titantic,palette='magma') 
```

```python
'''
3.pointplot()
折线图 - 置信区间估计 

'''
sns.pointplot(x='time',y='total_bill',hue='smoker',data=tips, 
             palette='hls',
              dodge=True ,#设置点是否分开 
              join=True, #是否连线 
              markers=['o','x'],0
              linestyles=['-','--'],#设置点样式、线形               
             )
tips.groupby(['time','smoker']).mean()['total_bill']
#计算数据
# 用法和barplot相似 
```
