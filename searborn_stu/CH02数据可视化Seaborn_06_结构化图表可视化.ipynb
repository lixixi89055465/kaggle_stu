```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
sns.set_style('darkgrid') 
sns.set_context('paper') 
# 设置风格、尺度 
import warnings
warnings.filterwarnings('ignore') 
# 不发出警告
```

```python
'''
1。基本设置 
绘制直方图 
'''
tips= pd.read_csv('tips.csv') 
tips.head()
# 导入数据 
```

```python
g=sns.FacetGrid(tips,col='time',row='smoker') 
# 创建一个绘图表格区域，设置好row,col 并分组 

g.map(plt.hist,'total_bill',alpha=0.5,color='k',bins=10) 
```

```python
'''
1.基本设置 
绘制直方图 
'''
g=sns.FacetGrid(tips, 
               col='day',
            size=4, # 图表大小 
            aspect=1 ,#图表长宽比 
               )
g.map(
     plt.hist,'total_bill',
       alpha=0.5,
       bins=10,
       histtype='step',# bar , barstacked, step, stepfilled 
       color='k',
      ) 
```

```python
'''
1.基本设置 
绘制散点图 
'''
g= sns.FacetGrid(tips,col='time',row='smoker') 
# 创建一个绘图表格区域，设置好row,col并分组 
g.map(plt.scatter, 
     'total_bill','tip',# share{x,y}-> 设置x,y数据 
      edgecolor='w',s=40,linewidth=1) # 设置点大小，描边宽度及颜色 
g.add_legend() 
# 添加图例 
```
